```python
f = '/home/cruman/Data/CONUS_I/CONUSI@Stations'
#provinces = ['AB', 'BC', 'MB', 'NB', 'NL', 'NS', 'ON', 'PE', 'QC', 'SK']
provinces = next(os.walk(f'{f}'))[1]
datei = datetime(2000,10,1,0)
datef = datetime(2013,9,30,23)

#arq = open('stations_dist_v2.dat', 'w')
#arq.write('CITY;LAT;LON;PEAK1;PEAK2;NEAR0;NEAR0_P01;NEAR0_P02;MEAN;STD\n')
time = []
di = datetime(2000,10,1,0,0)

sim = 'CTRL'
#sim = 'PGW'

for pv in provinces:
    
    # Empty dataframe to populate later
    df = pd.DataFrame({})
    # Folder of the netcdf files
    ff_CTRL = f"{f}/{pv}/{sim}"
    filename = glob(f"{ff_CTRL}/*nc")[0]
    
    # Name of the file, and removing the spaces and the extension
    fname = next(os.walk(f'{ff_CTRL}'))[2][0].replace(' ','_')[:-3] + f'_{sim}.csv'
    print(filename)
    
    # Opening the netcdf file
    xdata = xr.open_dataset(filename, engine='netcdf4')
    
    # Creating the time dimension, as the original dimension is int
    time = []
    for item in xdata.Time:
        time.append(di + relativedelta(hours=int(item.values))) 
    
    # The data is 4 points around the lat/lon of the station. 
    # Averaging it to just one point
    pr = xdata.PREC_ACC_NC.mean(dim=['south_north','west_east'])
    sn = xdata.SNOW_ACC_NC.mean(dim=['south_north','west_east'])
    psfc = xdata.PSFC.mean(dim=['south_north','west_east'])
    t2 = xdata.T2.mean(dim=['south_north','west_east'])
    u10 = xdata.U10.mean(dim=['south_north','west_east'])
    v10 = xdata.V10.mean(dim=['south_north','west_east'])
    rh2 = xdata.RH2.mean(dim=['south_north','west_east'])
    td2 = xdata.Td2.mean(dim=['south_north','west_east'])
    #xdata_mean = xdata.mean(dim=['XLAT','XLONG'])
    
    # populating the dataframe
    df['Time'] = time
    df[f'PR ({xdata.PREC_ACC_NC.units})'] = pr
    df[f'SN ({xdata.SNOW_ACC_NC.units})'] = sn
    df[f'PSFC ({xdata.PSFC.units})'] = psfc
    df[f'T2 ({xdata.T2.units})'] = t2
    df[f'U10 ({xdata.U10.units})'] = u10
    df[f'V10 ({xdata.V10.units})'] = v10
    df[f'TD2 ({xdata.Td2.units})'] = td2
    
    df.to_csv(fname)
    
    
    
```
